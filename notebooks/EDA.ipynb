## Initial Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('..\data\COVID DEATHS.csv')
```

```python
df.describe()
```

#### Each country reports covid data, new infections, deaths, ICU patients, starting from 3 JAN 2020 ending 12 APR 2023

#### Only looking at specific countries of interest, I have chosen to filter my dataframe.

```python
#Code to filter dataframe
countries_of_interest = ['Brazil', 'China', 'United States', 'Russia', 'India', 'Indonesia', 'Pakistan', 'Singapore', 'Germany', 'Mexico', 'Italy', 'Israel']
```

```python
filtered_df = df[df['location'].isin(countries_of_interest)]
```

```python
filtered_df.head()
```

### Develop line chart to compare countries according to new_cases_smoothed column.

```python
# Code to isolate the countries I want and chart them on a single plot

# Assuming you already have the filtered DataFrame named `filtered_df`
filtered_df['date'] = pd.to_datetime(filtered_df['date'])  # Convert 'date' column to datetime format
filtered_df = filtered_df.set_index('date')  # Set 'date' as the DataFrame's index

#countries_to_plot = ['Brazil', 'China', 'United States', 'Singapore', 'Russia', 'India', 'South Africa', 'Indonesia', 'Pakistan', 'Nigeria', 'Bangladesh', 'Mexico']
countries_to_plot = ['Brazil', 'China', 'United States', 'Russia', 'India', 'Indonesia', 'Pakistan']
column_to_plot = 'new_cases_smoothed_per_million'

```

```python
#Plot

for country in countries_to_plot:
    country_data = filtered_df[filtered_df['location'] == country]
    plt.plot(country_data.index, country_data[column_to_plot], label=country)

# Set plot title and labels
plt.title('New Cases per Million Weekly Average')
plt.xlabel('Date')
plt.ylabel('New Cases per Million')

# Customize the x-axis tick labels to show every week
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(15))  # Max 20 tick labels on the x-axis

# Rotate the x-axis tick labels for better visibility
plt.xticks(rotation=45)

# Add a legend to distinguish the lines
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Now do the same given these countries but focusing on the New Deaths Smoothed Column

countries_to_plot = ['Brazil', 'China', 'United States', 'Russia', 'India', 'Indonesia', 'Pakistan']
column_to_plot = 'new_deaths_smoothed_per_million'

for country in countries_to_plot:
    country_data = filtered_df[filtered_df['location'] == country]
    plt.plot(country_data.index, country_data[column_to_plot], label=country)

# Set plot title and labels
plt.title('New Deaths per Million Weekly Average')
plt.xlabel('Date')
plt.ylabel('New Deaths per Million')

# Customize the x-axis tick labels to show every week
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(15))  # Max 20 tick labels on the x-axis

# Rotate the x-axis tick labels for better visibility
plt.xticks(rotation=45)

# Add a legend to distinguish the lines
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

# Now seeking to merge my deaths .csv with the vaccine.csv into one data frame to use for further EDA.  There is a lot more country specific information including HDI, diabetes rates, population density, etc.


```python
df_vaccines = pd.read_csv('../data/COVID_VACCINATIONS.csv')
```

```python
df_vaccines.info()
```

```python
df_deaths = pd.read_csv('../data/COVID DEATHS.csv')
df_deaths.info()
```

```python
# Merging my two dataframes:

# Merging along the indices
merged_df = pd.merge(df_deaths, df_vaccines, left_index=True, right_index=True)
merged_df.info()
```

```python
columns_to_drop = ['iso_code_y', 'continent_y', 'location_y', 'date_y',
    'total_cases', 'new_cases', 'total_deaths', 'new_deaths',
    'total_cases_per_million', 'new_cases_per_million',
    'total_deaths_per_million', 'new_deaths_per_million',
    'icu_patients', 'hosp_patients', 'weekly_icu_admissions', 'weekly_hosp_admissions',
    'total_tests', 'new_tests', 'total_tests_per_thousand', 'new_tests_per_thousand',
    'new_tests_smoothed', 'new_tests_smoothed_per_thousand',
    'positive_rate', 'tests_per_case', 'tests_units',
    'total_boosters', 'new_vaccinations', 'new_vaccinations_smoothed',
    'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred',
    'total_boosters_per_hundred', 'new_vaccinations_smoothed_per_million',
    'new_people_vaccinated_smoothed', 'new_people_vaccinated_smoothed_per_hundred',
    'stringency_index', 'median_age', 'handwashing_facilities',
    'excess_mortality_cumulative_absolute', 'excess_mortality_cumulative']
main_df = merged_df.drop(columns = columns_to_drop).copy()
```

```python
main_df.info()
```

# Now let's do some EDA on the countries according to vaccination rates:

```python
#Now do the same given these countries but focusing on vaccination rates

countries_to_plot = ['Brazil', 'China', 'United States', 'Russia', 'India', 'Indonesia', 'Pakistan']
column_to_plot_a = 'people_fully_vaccinated_per_hundred'

for country in countries_to_plot:
    country_data = main_df[main_df['location_x'] == country]
    plt.plot(country_data['date_x'], country_data[column_to_plot_a], label=country)

# Set plot title and labels
plt.title('People Fully Vaccinated Per Hundred')
plt.xlabel('Date')
plt.ylabel('Number Fully Vaccinated Per Hundred')

# Customize the x-axis tick labels to show every week
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(15))  # Max 20 tick labels on the x-axis

# Rotate the x-axis tick labels for better visibility
plt.xticks(rotation=45)

# Add a legend to distinguish the lines
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Filter data to include only the desired date range
start_date = '2/7/2020'
end_date = '5/23/2023'
main_df_filtered = main_df[(main_df['date_x'] >= start_date) & (main_df['date_x'] <= end_date)]

countries_to_plot = ['Brazil', 'China', 'United States', 'Russia', 'India', 'Indonesia', 'Pakistan']
column_to_plot_a = 'people_fully_vaccinated_per_hundred'

for country in countries_to_plot:
    country_data = main_df_filtered[main_df_filtered['location_x'] == country]
    plt.plot(country_data['date_x'], country_data[column_to_plot_a], label=country)

# Set plot title and labels
plt.title('People Fully Vaccinated Per Hundred')
plt.xlabel('Date')
plt.ylabel('Percentage of People Fully Vaccinated')

# Customize the x-axis tick labels to show every week
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(15))  # Max 20 tick labels on the x-axis

# Rotate the x-axis tick labels for better visibility
plt.xticks(rotation=45)

# Add a legend to distinguish the lines
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
